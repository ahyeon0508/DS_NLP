# Doc2Vec과 추천 시스템을 이용한 대학 학과 추천 프로젝트

- 캡스톤종합디자인프로젝트 논문 후보로 해당 미니 프로젝트를 진행함
- 대입 서류를 통해 대학 학과를 추천 받음으로써 아직 학과를 선택하지 못 한 학생들에게 도움을 줄 수 있음  
- 또한 현재 대한민국에 존재하는 대학과 학과의 수가 매우 많기 때문에 학과 선택의 폭을 넓혀줄 수 있음
- 사용 데이터 : 커리어넷 오픈 API 학과정보

-----

## 데이터 호출 및 저장

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
def util(url):
    request = urllib.request.urlopen(url)
    xml = request.read()
    soup = BeautifulSoup(xml, "html.parser")

    return soup
```

```python
# 인증키 입력
apiKey = "본인 인증키 입력"
```

```python
def collect_info():
    url = "http://career.go.kr/cnet/openapi/getOpenApi?apiKey=" + apiKey + "&svcType=api" \
          "&svcCode=MAJOR&contentType=xml&gubun=univ_list"

    soup = util(url)
    
    # 전체 검색 결과수를 통해 모든 학과코드 가져오기
    totalCount = soup.find("totalcount").text

    url = "http://career.go.kr/cnet/openapi/getOpenApi?apiKey=" + apiKey + "&svcType=api" \
          "&svcCode=MAJOR&contentType=xml&gubun=univ_list&perPage=" + totalCount

    soup = util(url)
    major = soup.find_all("content")

    # 학과코드
    major_code = []
    
    for code in major:
        major_code.append(code.find("majorseq").text)
        
    # 학과 정보 모으기
    summary = []
    interest = []
    major_property = []
    university = []
    
    for major_num in major_code:
        major_url = "http://career.go.kr/cnet/openapi/getOpenApi?apiKey=" + apiKey + "&svcType=api" \
          "&svcCode=MAJOR_VIEW&contentType=xml&gubun=univ_list&majorSeq=" + major_num
        major_soup = util(major_url)
        
        summary.append(major_soup.find("summary").text)
        interest.append(major_soup.find("interest").text)
        major_property.append(major_soup.find("property").text)
        univ_content = major_soup.find("university").find_all("content")
        
        university_data = []
        
        for data in univ_content:
            university_data.append(data.find("majorname").text + "(" + data.find("schoolname").text + ") ")
            
        university.append(university_data)
                
    df_major = pd.DataFrame({"학과개요" : summary, "흥미와적성" : interest,
                         "학과특성" : major_property, "대학" : university})
    
    return df_major
```

```python
df_major = collect_info()
```
