# KR-WordRank
- '새내기路(로) : AI기반 대입 모의 면접 솔루션' 프로젝트 기능 중 하나인 학교생활기록부, 자기소개서를 분석하여 질문을 추출하는 기능에 키워드 추출이 필요함
- 따라서 WorkRank에 한국어의 특징을 반영한 KR-WordRank을 테스트 해봄.

-------

## 라이브러리 호출

```python
from krwordrank.word import KRWordRank
from krwordrank.hangle import normalize
from krwordrank.word import summarize_with_keywords
```

## 데이터 읽기 및 전처리

```python
fname = 'data/생활기록부.txt'
sentences = []

with open(fname, encoding="utf-8") as f:
    for _ in range(3):
        sentence = next(f).strip()
        sentence = normalize(sentence, english=True, number=True) # 단어 추출에 영어/숫자를 포함하기 위해
        sentences.append(sentence)
print(sentences)
```

* 특수문자 전처리

```python
for i in range(len(sentences)):
    sentences[i] = sentences[i].replace(".", "").replace(",", "").replace("(", "").replace(")", "").replace("’", "").replace("‘", "")

print(sentences)
```

## 학습 및 키워드 출력

```python
keywords = summarize_with_keywords(sentences, min_count=2, max_length=10,
    beta=0.85, max_iter=10, verbose=True) # verbose는 학습의 진행상황을 보여주는 것임

for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:30]:
    print('%8s:\t%.4f' % (word, r))
```

## Stopwords

- 생활기록부에 공통적으로 나올 단어를 분석자 주관으로 제거함

```python
stopwords = {'등의', '대한', '많아', '학생임', '있으며', '또한', '친구들', '통해', '실시함', '보이는', '문제를'}

keywords = summarize_with_keywords(sentences, min_count=2, max_length=10,
    beta=0.85, max_iter=10, stopwords=stopwords, verbose=True)

for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:20]:
    print('%8s:\t%.4f' % (word, r))
```

## 키워드 정리

- 조사 등을 제거하고 싶음 : KoNLPy 사용

```python
from konlpy.tag import Okt

okt = Okt()

for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:20]:
    print('%8s:\t%.4f' % (okt.pos(word), r))
```

- 명사, 동사, 형용사만 추출

```python
for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:20]:
    preprocessing_words = []
    for word_pos in okt.pos(word):
        if word_pos[1] in ['Noun', 'Verb', 'Adjective']:
            preprocessing_words.append(word_pos[0])
    final_word = ' '.join(preprocessing_words)
    print(final_word)
```
